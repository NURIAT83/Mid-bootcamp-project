Dickey-Fuller test (unit root)
We conduct the following test in our main variables to know if our data is stationary or not.
p-value > 0.05: Fail to reject the null hypothesis (H0), the data has a unit root and is non-stationary.
p-value <= 0.05: Reject the null hypothesis (H0), the data does not have a unit root and is stationary.

```python
Hindex0 = numeric_CR['H index'] 
```

```python
from statsmodels.tsa.stattools import adfuller
X = Hindex0[1:].values

result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

Taking into consideration the p-value (0.00) of the variable H-index0 , we reject the null hypothesis, the data does not have unit root and is stationary.

```python
Hindex1 = numeric_CR['H index'] - numeric_CR['H index'].shift(1)
```

```python

from statsmodels.tsa.stattools import adfuller
X = Hindex1[1:].values

result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

Taking into consideration the p-value (0.00) of the variable H-index1 lag(1), we reject the null hypothesis, the data does not have unit root and is stationary.

We conclude that the variable H-index does not present seasonality.
